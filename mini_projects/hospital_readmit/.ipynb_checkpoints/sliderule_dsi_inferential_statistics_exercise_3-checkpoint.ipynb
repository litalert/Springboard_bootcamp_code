# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# Cleaning data 

df = clean_hospital_read_df[['Number of Discharges', 'Excess Readmission Ratio']]
df = df.dropna()
```

```python
# Investigate the data

med = np.median(df['Number of Discharges'])

# Splitting the data into 2 groups- high capacity and low capacity hospitals
high = df[df['Number of Discharges'] >= med].reset_index()
low = df[df['Number of Discharges'] < med].reset_index()
high.columns = ['index', 'num_discharges', 'readmission']
low.columns = ['index', 'num_discharges', 'readmission']

# Calculating both groups' readmission rate means
m_high = np.mean(high['readmission'])
m_low = np.mean(low['readmission'])
diff = np.float64(m_low - m_high)

print('Result 1')
print('\n')
print('m_high:', m_high)
print('m_low:', m_low)
print('diff:', diff)
```

```python
# Performing two sample bootstrap hypothesis testing

# H0- There is no difference in mean excess readmission rates in high and low capacity hospitals (m_low - m_high = 0)
# H1- There is a difference in mean excess readmission rates in high and low capacity hospitals (m_low != m_high)

# Shifting the means
mean = (m_high + m_low)/2
high_shifted = high.readmission + mean - m_high
low_shifted = low.readmission + mean - m_low

bs_diff = []
for i in range(100000):
    bs_high = np.random.choice(high_shifted, size=len(low))
    bs_low = np.random.choice(low_shifted, size=len(low))
    bs_high_m = np.mean(bs_high)
    bs_low_m = np.mean(bs_low)
    bs_diff.append(bs_low_m - bs_high_m)
p_value = np.sum(bs_diff >= diff)/len(bs_diff)

print('Result 2')
print('\n')
print('p-value:', p_value)
print('Based on α = 0.01, we are rejecting the null hypothesis of there is no statistical differences in readmission ratio between low and high capacity hospitals (m_low - m_high = 0)')

plt.hist(bs_diff, bins=int(np.sqrt(len(bs_diff))))
plt.xlabel('Difference in Excess Readmission Ratio')
plt.title('Sampling Distribution of the Difference in Excess Readmission Ratio Between Low and High Capacity Hospitals')
plt.show()
```

```python
# Plotting box plot
df = df.reset_index()
for i in df.index:
    if df.loc[i, 'Number of Discharges'] <=280:
        df.loc[i, 'group'] = '0-280'
    elif df.loc[i, 'Number of Discharges'] <=1000:
        df.loc[i, 'group'] = '501-1000'
    elif df.loc[i, 'Number of Discharges'] <=1500:
        df.loc[i, 'group'] = '1001-1500'
    elif df.loc[i, 'Number of Discharges'] <=2000:
        df.loc[i, 'group'] = '1501-2000'
    elif df.loc[i, 'Number of Discharges'] <=2500:
        df.loc[i, 'group'] = '2001-2500'
    else:
        df.loc[i, 'group'] = float('nan')
      
sns.boxplot(x='group', y='Excess Readmission Ratio', data=df, whis=2)
#sns.swarmplot(x='group', y='Excess Readmission Ratio', data=df)
plt.title('Number of Discharges VS Excess Rate of Readmission')
plt.xlabel('Number of Discharges')
plt.show()
```
