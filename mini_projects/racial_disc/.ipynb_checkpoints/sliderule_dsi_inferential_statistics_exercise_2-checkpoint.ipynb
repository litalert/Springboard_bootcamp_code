```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from random import shuffle
```

```python
# Importing data
data = pd.io.stata.read_stata('data/us_job_market_discrimination.dta')
data.head()
```

```python
# Calculating success (call for resume) porportion for black and white resumes

b = sum(data[data.race=='b'].call)
w = sum(data[data.race=='w'].call)

b_tot = len(data[data.race == 'b'])
w_tot = len(data[data.race == 'w'])

diff = (w/w_tot) - (b/b_tot)
p_B = b/b_tot
p_W = w/w_tot

print('Result 1')
print('p_B:', p_B)
print('p_W:', p_W)
print('diff:',diff)
```

```python
# Performing two sample bootstrap hypothesis testing

# Calculating mean proportion and simulating list of 1's and 0's to reflect the mean proportion
ones = (b + w)/2
zeros = int(b_tot - ones)

sample = np.ones(int(ones), dtype=np.int)
sample = sample.tolist()
for i in range(zeros):
    sample.append(0)
shuffle(sample)
 
# Creating boostrap samples, compute ratios, and compute difference 
bs_diff = []
for i in range(100000):
    bs_b = np.random.choice(sample, size=b_tot)
    bs_w = np.random.choice(sample, size=w_tot)
    bs_p_b = np.sum(bs_b)/len(bs_b)
    bs_p_w = np.sum(bs_w)/len(bs_w)
    bs_diff.append(bs_p_w - bs_p_b)

p_value = np.sum(bs_diff >= diff)/len(bs_diff)
print('Result-2')
print('\n')
print('H0- p_W - p_B = 0')
print('H1- p_W != p_B')
print('p-value: ', p_value)
print('We are rejecting the null hypothesis of there is no difference in success proportion between black and white resumes')

plt.hist(bs_diff, bins=int(np.sqrt(len(bs_diff))))
plt.title('Sampling Distribution of the Difference in Success Proportion')
plt.xlabel('Difference in Success Proportion')
plt.show()
```

```python
# Computing confidence interval and margin of error

# Calculating standard error of the differenced samples
std_err = np.sqrt((np.sqrt((p_W*(1-p_W))/b_tot))**2 + (np.sqrt((p_B*(1-p_B))/w_tot))**2)

# Calculating standard error based on the simulated distribution (sanity check only)
#std_err_bs = np.std(bs_diff)

# Assuming 95.4% confidence level
margin_err = round(std_err*2, 2)
lower = round(diff-margin_err, 2)
upper = round(diff+margin_err, 2)

print('Result-3')
print('\n')
print('We are 95% confident that the true difference in success proportion is between', lower, 'and', upper)
```

```python
# Investigating difference in success (callback) ratio based on other factors 

# yearsexp
more = sum(data[data.yearsexp >= 7].call)
less = sum(data[data.yearsexp < 7].call)
tot_m = len(data[data.yearsexp >= 7])
tot_l = len(data[data.yearsexp < 7])

print('Result-4')
print('\n')
print('yearsexp:', more/tot_m, less/tot_l)

# honors
more = sum(data[data.honors >= 1].call)
less = sum(data[data.honors < 1].call)
tot_m = len(data[data.honors >= 1])
tot_l = len(data[data.honors < 1])

print('honors:', more/tot_m, less/tot_l)
```
